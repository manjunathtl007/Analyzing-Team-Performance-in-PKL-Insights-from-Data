# Determine the which franchise is most successful in the PKL history

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('C:\\final_dataset_pkl.csv')
```

```python
df
```

# Most number of matches won by each team

```python
df = df.sort_values('Won', ascending=False)
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(35, 7))
plt.bar(df['Name'], df['Won'])
for a, b in zip(df['Name'],df['Won']):
    plt.text(a, b, str(b), fontsize=15)
plt.title('Total_win', fontsize = 25, color = 'r')
plt.xlabel('Teams',fontsize = 20)
plt.ylabel('No. of wins',fontsize = 20)
plt.show()
```

# Most number of Trophies

```python
df = df.sort_values('Trophy', ascending=False)
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(35, 7))
plt.bar(df['Name'], df['Trophy'])
for a, b in zip(df['Name'],df['Trophy']):
    plt.text(a, b, str(b), fontsize=15)
plt.title('No. of Champions', fontsize = 25, color = 'r')
plt.xlabel('Teams',fontsize = 20)
plt.ylabel('No. of Trophies',fontsize = 20)
plt.show()
```

# Most number of matches loss by each team

```python
df = df.sort_values('Loss', ascending=False)
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(35, 7))
plt.bar(df['Name'], df['Loss'])
for a, b in zip(df['Name'],df['Loss']):
    plt.text(a, b, str(b), fontsize=15)
plt.title('Total_losses_ by_each_team', fontsize = 25, color = 'r')
plt.xlabel('Teams',fontsize = 20)
plt.ylabel('No. of losses',fontsize = 20)
plt.show()
```

# Total Raid Points by each team

```python
df = df.sort_values('Raid_Points', ascending=False)
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(35, 7))
plt.bar(df['Name'], df['Raid_Points'])
for a, b in zip(df['Name'],df['Raid_Points']):
    plt.text(a, b, str(b), fontsize=15)
plt.title('Total_Raid_Points_ by_each_team', fontsize = 25, color = 'r')
plt.xlabel('Teams',fontsize = 20)
plt.ylabel('No. of Raid Points',fontsize = 20)
plt.show()
```

# Total Tackle Points by each team

```python
df = df.sort_values('Tackle_Points', ascending=False)
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(35, 7))
plt.bar(df['Name'], df['Tackle_Points'])
for a, b in zip(df['Name'],df['Tackle_Points']):
    plt.text(a, b, str(b), fontsize=15)
plt.title('Total_Tackle_Points_ by_each_team', fontsize = 25, color = 'r')
plt.xlabel('Teams',fontsize = 20)
plt.ylabel('No. of Tackle Points',fontsize = 20)
plt.show()
```

# From the above data analysis we can conclude that the Patna Pirates Team is the most successful franchise in the PKL history
